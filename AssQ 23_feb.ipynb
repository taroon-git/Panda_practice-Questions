Q:7 Create a DataFrame using multiple Series. Explain with an example

```python
import pandas as pd

r = pd.Series()
for i in range(0, 10):
    r = r.set_value(i,i*3)
r.name = 'r1'

s = pd.Series()
for i in range(0, 10):
    s = s.set_value(i,i*5)
s.name = 's1'

pd.DataFrame({'r': r, 's': s})
```
